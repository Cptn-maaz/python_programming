Set: defined with curly braces {} or set() function. Unordered, does not allow duplicates, mutable

```python
# Creating set
empty_set = set()

number_set = {1,2,3,2,4,5}    # sets do not allow duplicate values, so 2 will be stored only once in the set

mixed_set = {1, 3.4, "Hello"}

collection_set = {(1,2), "Great", 10}

print("Empty set: ", empty_set)
print("Number set: ", number_set)
print("Mixed set: ", mixed_set)
print("Collection set: ", collection_set)
```

```python
# Adding items to a set
print("Before adding an element: ", mixed_set)
mixed_set.add("welome")
print("After adding and element: ", mixed_set)
```

```python
# Removing items from a set
print("Before removing elements: ", mixed_set)
mixed_set.discard(6) # discard() will remove the specified element from the set, if it is present in the set, if not present it will do nothing and will not raise an error
mixed_set.remove(3.4) # remove() will remove the specified element from the set, if it is present in the set, otherwise it will raise a KeyError
print("After removing elements: ", mixed_set)
```

```python
# applying union, intersection and difference on sets

set1 = {1,2,3,4}
set2 = {3,4,6,7}

union_set = set1.union(set2) # will return a new set that contains all the unique elements from both sets
print("Union of set1 and set2: ", union_set)

intersection_set = set1.intersection(set2) # will return a new set that contains only the common elements in both sets
print("Intersection of set1 and set2: ", intersection_set)

difference_set1 = set1.difference(set2) # will return a set that contains the elements that are present in set1 but not in set2
print("Difference of set1 and set2: ", difference_set1)

difference_set2 = set2.difference(set1) # will return a set that contains the elements that are present in set2 but not inset1
print("Difference of set2 and set1; ", difference_set2)


#  Symmetric_diiference() will return a set that contains the elements that are present in either set1 or set2 but not in both sets i.e not common elements
symmetric_diff = set1.symmetric_difference(set2) 
print("Symmetric difference of set1 and set2: ", symmetric_diff)
```

Operators: Performs computations and logical operations on data

```python
# Assignment operator(=): Assign values to variables
car = "BMW"      # value(BMW) has been assigned to variable(car)
car_price = 20000000  # value(20000000) has been assigned to variable(car_price)

print("Car Name: ", car)
print("Car Price: ", car_price)
```

```python
# Arithmetic operators(+,-,*,/,% etc): Performs mathematical operations like addition, subtraction etc.

number1, number2 = 10, 3
print("Addition: ", number1 + number2)  # (+) returns addition
print("Subtraction: ", number1 - number2) # (-) returns subtraction
print("Multiplication: ", number1 * number2) # (*) returns multiplication
print("Division: ", number1 / number2) # (/) Returns quotient with decimals
print("Reminder: ", number1 % number2) # (%) Returns Reminder
print("Floor Division: ", number1 // number2) # (//) Returns quotient without any decimals
print("Suare of number1: ", number1 ** number2) # (**) is used to take exponent of a number
```

```python
# Comparison operators(>,<,>=,<=,!= etc): Compares two or more values, and will return answer in form of (True, False)

price1, price2 = 2000, 40000
print("price1 : ", price1)
print("price2 : ", price2)
print("Is price1 greater than price2? : ", price1 > price2) # (>) used for greater than
print("Is price1 less than price2? : ", price1 < price2) # (<) used for less than
print("Is price1 equal to price2? : ", price1 == price2) # (==) used for equality
print("Is price1 not equal to price2? : ", price1 != price2)# (!=) used for not equal
print("Is price1 greater than or equal to price2? : ", price1 >= price2) # (>=) used for greater than or equal to
print("Is price1 less than or equal to price2? : ", price1 <= price2) # (<=) used for less than or equal to
```

```python
# Logical Operators(and, or, not): Used to combine multiple conditions togather

condition1, condition2 = True, False
print("condition1 : ", condition1)
print("conditoin2 : ", condition2)
print("Are both conditions true? : ", condition2 and condition2) # (and) returns true if both cnditions are true.
print("Is either of the condition true? : ", condition1 or condition2) # (or) returns true if either of a condition is true
print("Is condition1 not true? : ", not condition1) # (not) returns true if the condition is not true
print("Is condition2 not ture? : ", not condition2)
```

```python
# Membership operators (in, not in): Used to check if a value or variable exists in a collection(list, tuple, string etc) or not
# It will output in boolean (True, False)
name = "Mauz Ahmad"
print("n" in name) # checks if character "n" is present in value of variable "name" or not.
print("M" in name) # checks if character "M" is present in value of variable "name" or not.
print("o" not in name)
```

Loops: Used to execute some specific lines of code as many times as we need.

For Loop: Used for iterating over a sequence  i.e (string, list, tuple, dictionary, set etc). it process each element one by one at a time in a readable format.

```python
# iterating a list containing numbers using for() loop
numbers = [1,3,4,6,2,7]  # list of numbers
for num in numbers:    # iterates each number in the above list and save it in the iterator(num)
    print(num)   # prints the iterator(num)
```

```python
# iterating a string
name = "Muhammad Maaz"  # string
for chars in name:  # iterates each character in the string and save it in the iterator (chars)
    print(chars)    # prints the iterator (cahrs) every time.
```

```python
# enumerate() function in for() loop: Through enumerate() we can iterate values with indxes in a collection (list, tuple etc)
vegetables = ["onion", "garlic", "potato"] # list of vegetables
for index, vegetable in enumerate(vegetables): # iterate indexes, saves them in iterator "index". and iterate vegetables and save them in iterator (vegetable)
    print(f"Vegetable on index {index} : {vegetable}") # prints both the iterators (index and vegetable)
```

```python
# performing the above task without enumerate() function
fruits = ["Apple", "Banana", "Mango"]

index = 0  # initialized a counter variable

for fruit in fruits: # iterates elements in fruits and saves them in the iterator (fruit)
    print(f"Fruit on index {index} : {fruit}")  # prints counter (index) and iterator (fruit)
    index += 1    # incrementing index value manually i.e index = index +1
```

```python
# iterating a dictionary using for() loop
information = {
    "name" : "Muhammad Talha",     # key = "name" , value = "Muhammad Talha"
    "Age" : 9,                     # key = "Age"  , value = 9
    "Height" : "110cm"             # key = "Height", value = "110cm"
}
# iterating over keys
print("iterating over keys:")
for key in information:      # iterate each key in dictionary and saves it in the iterator (key)
    print(f"{key} : {information[key]}") # prints the iterator (key) and also access values using the key
```

```python
# performing the above task using keys values pairs
print("iterating over key-value pairs")
for key, value in information.items():
    print(f"{key} : {value}")
```

```python
# iterating tuples using for()loop
gen_tup = ("Maaz", 27, "165cm", 0.5) # tuple containing mixed data
for data in gen_tup:
    print(data)
```

```python
# iterating the above values with index using for() loop on a tuple
for index, value in enumerate(gen_tup):   # iterate indexes and values and saves it in iterators (index) and (value) respectively
    print(f"value on index {index} : {value}") # printing indexes and its values side by side
```

```python
# performing the above task with out using enumerate() function
ind_counter = 0  # initializing a counter variable for indexes
for value in gen_tup:
    print(f"value on index {ind_counter} : {value}")
    ind_counter += 1
```

```python
# iterating sets using for() loop
info_set = {"Abuzar", "170cm", 22} # tuple with mixed data
for info in info_set: # iterates each value and saves it in the iterator (info)
    print(info) # prints the iterator (info)
```

```python
# iterating the above set values with indexes
for ind_set, value in enumerate(info_set): # iterates indexes and values and saves it in interators (ind_set and value) respectively
    print(f"vlaue on index {ind_set} : {value}") # prints the iterators (ind_set and value) side by side
```

```python
# performing the above task without using enumerate() function
set_ind_counter = 0 # initializes a counter valriable for indexes
for value in info_set: # iterates each value in info_set and saves it in iterator (value)
    print(f"value on index {set_ind_counter} : {value}") # prints indexes (set_ind_counter) and inerator (value) side by side
    set_ind_counter += 1 # incrementing index value manually i.e set_ind_counter = set_ind_counter + 1
```
