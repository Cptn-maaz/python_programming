A program that asks for a traffic light color (red, yellow, green) and prints the correct action. <br>
- Green ---> "Go!!"
- Yellow ---> "Slow down and prepare to Stop"
- Red ---> "Stop!!"

```python
# lets write the program
traffic_signal = input("Enter the traffic signal: ") # takes input from user and defines traffic signal
if traffic_signal == "Green" or traffic_signal == "green": # checks the first conditon
    print(f"Signal: {traffic_signal} ----> Go!!") # if signal is green, first condition will True and this block will execute

elif traffic_signal == "Yellow" or traffic_signal == "yellow": # checks the second condition if the first condtion id "False"
    print(f"Signal: {traffic_signal} -----> Slow down and prepare to Stop") # if signal is yellow, this block will execute

else: # if both the above conditions (if, elif) are wrong, its means that the light is definitely red, so this (else) block will execute
    print(f"Signal: {traffic_signal} ----> Stop!!")
```

A program that asks the users for their age and checks if they are eligible to join the basketball team or not. <br>
Age must be 16 or greater that 16

```python
player_age = int(input("Enter your age: ")) # takes age (integer) as input and defines age of the player
if player_age >= 16: # checks if the player_age is equal or greater than 16
    print(f"Player Age: {player_age} ----> Welcome! You can join the team.") # if the above condition is "True", this block will execute

else: # if the above (if) condition is "False", this else block will execute
    print(f"Player Age: {player_age} ----> Sorry! You are not eligible")
```

A program that takes temperature in &deg;c and prints the weather status according the the following conditions. <br>
- Above 30&deg;c ----> "It's hot outside!"
- Between 15&deg;c and 30&deg;c ----> "It's a pleasent weather"
- Below 15&deg;c ----> "It's cold outside!"

```python
temperature = float(input("Enter the temperature: ")) # Defines temperature
if temperature >= 30: # checks if temoerature is high i.e (equal or greater than 30)
    print(f"Temperature: {temperature}℃ ----> It's hot outside!") # if the above condition is "True", this block will execute

elif temperature > 15 and temperature < 30: # if the first condition (if) is "False", then this 2nd condition will be checked
    print(f"Temperature: {temperature}℃ ----> It's a pleasent weather") # if the 2nd (elif) condition is "True", this block will execute

else: # if both the above conditions are "False", then definitely it will be the third scenario, and hence this (else) block will execute
    print(f"Temperature: {temperature}℃ ----> It's cold outside!")
```

A program that asks for a student's exam score (0 - 100) and assigns him or her a grade according to the following conditions. <br>
- 90-100 ----> "Grade A"
- 80-89 ----> "Grade B"
- 70-79 ----> "Grade C"
- 60-69 ----> "Grade D"
- Below 60 ----> "Grade F"

```python
student_score = int(input("Enter the student's score: ")) # defines student's score

if student_score >=90 and student_score <= 100: # check if score is in 90-100
    print(f"Student Score: {student_score} ----> Grade A , Congrates!") # executes if the above condition is "True"

elif student_score >= 80 and student_score < 90: # checks if score is un range 80-89
    print(f"Student Score: {student_score} ----> Grade B , Congrates!") # executes if the above (if) is "False" and this (elif) is "True"

elif student_score >= 70 and student_score < 80: # checks if score is in range 70-79
    print(f"Student Score: {student_score} ----> Grade C , Keep it up!") # ececutes if both the above (if, elif) are "False"

elif student_score >= 60 and student_score < 70: # checks if score is in range 60-69
    print(f"Student Score: {student_score} ----> Grade D , Keep it up!") # executes if all the above three (if, elif, elif) are "False"

else: # if all the above conditions fails, it's means that it is definitely the 4th case i.e score is Below 60, then this (else) will execute
    print(f"Student Score: {student_score} ----> Grade F , Better Luck next time")
```

A program that asks user for year and checks whether it is leap year or not. <br>
A leap yaer will be checked according to the following conditions
- If the year is not divisible by 4 → Not a leap year.
- If the year is divisible by 4 but not by 100 → Leap year.
- If the year is divisible by 100 but not by 400 → Not a leap year.
- If the year is divisible by 400 → Leap year.

```python
year = int(input('Enter year: ')) # Defines year

if year % 4 != 0: # Checks the first condition
    print(f"Year: {year} ----> Not a leap year")

elif year % 4 == 0 and year % 100 != 0: # checks the 2nd condition
    print(f"Year: {year} ----> Leaf year.")

elif year % 100 == 0 and year % 400 != 0: # checks the 3rd condition
    print(f"Year: {year} ----> Not a leap year.")


else: # if all the above conditions fails, then its definitely the 4th condition 
    print(f"Year: {year} ----> Leap year") # this block will execute if all the above failed
```

A program for an ATM cash Withdrawal system on the basis of the following conditions <br>
- Asks the user for a withdrawal amount
- Checks if the account balance is sufficient
- If balance is enough, deduct and print the new balance
- If not, print "Insufficient fund!"

```python
withdrawal_amount = int(input("Enter amount")) # defines the withdrawal amount
account_balance = 10000 # defines account balance

if withdrawal_amount <= account_balance: # checks if the balance is sufficient
    print(f"Amount entered: {withdrawal_amount}pkr")
    print("----------------- Transaction Successful ----------------------------")
    print(f"You have successfully withdrawn {withdrawal_amount}pkr")
    account_balance = account_balance - withdrawal_amount  # deducting the account balance
    print(f"Your new balance is: {account_balance}pkr") # displays new account balance after deduction

else: # if the above condition fails, this block will execute
    print(f"Amount entered: {withdrawal_amount}pkr")
    print("Insufficient fund! --------------")
```
