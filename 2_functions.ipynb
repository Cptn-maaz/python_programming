Function: A reusable block of code designed to perform a specific task. Functions are used <br>
- To avoid writing repeatitive code (a bock of code that is required multiple time in a program)
- Code reusability, modularity, easier maintenance, abstraction, debugging and testing

Function types: <br>
 Built-in Functions: These are already built-in in Python, we don't have to write it from scratch. i.e print(), type(), range(), list() etc.

 User defined functions: These are the functions that a users have to write according to their needs.

Lets try some built-in functions

```python
# print(): Displays anything written inside it to the screen
name = "Muhammad" # defines name
print(f"Hello {name}! Welcome to the group")
```

```python
# input(): Takes input from users at runtime
last_name = input("Enter your last name: ") # takes input form user and assigns it to variable "last_name"
print(f"Hello Muhammad {last_name}! welcome to the group")
```

```python
# type(): checks type of data i.e (int, float, str, list etc) stored in a variable.
greetings = "Good Moring" # defines variable "greetings"
num_sequence = [2,4,6,8,10] # defines variable "num_sequence"

print(greetings, type(greetings)) # will print data in "greetings" and its type.
print(num_sequence, type(num_sequence)) # will print data in "num_sequence" and its type
```

```python
# len(): Finds length of a variable ie (How many characters are saved in a variable)
wish = "Congrates" # defines variable "wish"
age = "26 years" # defines "age"
running = "1.6 km" # defines "running"

print(wish,",","Length: ", len(wish)) # prints data in "wish" and its length
print(age,",","Length: ", len(age)) # prints data in "age" and its length
print(running,",","Length: ", len(running)) # print data in "running" and its length


# Note: We can only apply len() functions on a collection or a sequence i.e (string, list, tuple etc)
# We cannot apply it on a sigle integer or float value
```

```python
# max(): Returns the Highest value in a sequence.
prices = [300, 500,200,100,150] # defines a list of prices
highest_price = max(prices) # finds the highest price
print(f"The highest price is: {highest_price}")
```

```python
# sorted(): arrange data by default in accending order in a sequence.
marks = [30, 20, 40, 80, 100, 75, 60, 90] # defines a list of marks
print(f"Unsorted marks: {marks}")
sorted_marks = sorted(marks) # sort marks in accending order
print(f"Sorted Marks: {sorted_marks}")
```

```python
# range(): Generates a sequence of numbers with in a given range.
for numbers in range(6): # generates 6 numbers, starts from 0 and ends on 6-1 = 5
    print(numbers)
```

```python
# range() can take upto 3 argumnets i.e
# range(start, stop, increment)
for number in range(3, 20, 2): # will start form 3, stops on 20-1 = 19, and will increment iterator by 2
    print(number)

# Note: if we did'nt mention incrementation, it will increment iterator (number) by 1 by default
```

```python
# id(): id() function returns memory address of a variable where it has been stored.
vehicle = "Car" # defines vehicle
vehicle_memory_address = id(vehicle) # returns memory address of variable "vehicle"
print(f"The memory address of vehicle is: {vehicle_memory_address}") 
```

```python
# eval(): evaluates mathemical operations inside a string
calculation = "2 + 3 * 4" # defines "calculation" with a string value
print(f"Without using eval(), the result is: {calculation}") # will display as at as in the string

evaluation = eval(calculation) # evaluates the string's mathematical values
print(f"With using eval(), the result is: {evaluation}") # will display the result of math operations inside string
```
